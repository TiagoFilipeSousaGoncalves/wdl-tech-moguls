```python

from pandas_ods_reader import read_ods
import pandas as pd

churn_od = pd.read_csv('../data/Churn_OD.txt', delimiter = "|", encoding= 'utf-8')

#county from main dataset
county=churn_od['County_of_Public_Transportation'].str.lower()

#Unemployment by idade for sep 2019 - February 2020 and sep 2020 - January 2021

mes=['setembro','outubro','novembro','dezembro','Janeiro','fevereiro']
dfn_1=[]
for idx,i in enumerate(mes):
    if i=='setembro'or i=='outubro' or i=='novembro'or i=='dezembro':
        ano=2019
        path = "../data/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 2
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,285,287])
        df = df.drop(columns=["unnamed.2","unnamed.3","unnamed.6","unnamed.8","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': '< 25 Anos', 'unnamed.7': '25 - 34 Anos', 'unnamed.9': '35 - 54 Anos', 'unnamed.11': '55 Anos e +', 'unnamed.12': 'TOTAL'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn_1.append(df)
    else:
        ano=2020
        path = "../data/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 2
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,285,287])
        df = df.drop(columns=["unnamed.2","unnamed.3","unnamed.6","unnamed.8","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': '< 25 Anos', 'unnamed.7': '25 - 34 Anos', 'unnamed.9': '35 - 54 Anos', 'unnamed.11': '55 Anos e +', 'unnamed.12': 'TOTAL'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn_1.append(df)

#unemployment dataframe creation for sep 2020 - January 2021
mes=['setembro','outubro','novembro','dezembro','Janeiro']
dfn_2=[]
for idx,i in enumerate(mes):
    if i=='setembro'or i=='outubro' or i=='novembro'or i=='dezembro':
        ano=2020
        path = "../data/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 2
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,285,287])
        df = df.drop(columns=["unnamed.2","unnamed.3","unnamed.6","unnamed.8","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': '< 25 Anos', 'unnamed.7': '25 - 34 Anos', 'unnamed.9': '35 - 54 Anos', 'unnamed.11': '55 Anos e +', 'unnamed.12': 'TOTAL'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn_2.append(df)
    else:
        ano=2021
        path = "../data/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 2
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,285,287])
        df = df.drop(columns=["unnamed.2","unnamed.3","unnamed.6","unnamed.8","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': '< 25 Anos', 'unnamed.7': '25 - 34 Anos', 'unnamed.9': '35 - 54 Anos', 'unnamed.11': '55 Anos e +', 'unnamed.12': 'TOTAL'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn_2.append(df)

        
df_temp=pd.concat(dfn_1[i] for i in range(6))
df_temp['< 25 Anos'] = df_temp['< 25 Anos'].astype(float)
df_temp['25 - 34 Anos'] = df_temp['25 - 34 Anos'].astype(float)
df_temp['35 - 54 Anos'] = df_temp['35 - 54 Anos'].astype(float)
df_temp['55 Anos e +'] = df_temp['55 Anos e +'].astype(float)
df_temp['TOTAL'] = df_temp['TOTAL'].astype(float)
df_unemployment_1=df_temp.groupby(['concelhos']).mean().astype(int).reset_index()
#df_unemployment_1=df_unemployment_1[df_unemployment_1['concelhos'].isin(county.unique())].reset_index()


df_temp=pd.concat(dfn_2[i] for i in range(4))
df_temp['< 25 Anos'] = df_temp['< 25 Anos'].astype(float)
df_temp['25 - 34 Anos'] = df_temp['25 - 34 Anos'].astype(float)
df_temp['35 - 54 Anos'] = df_temp['35 - 54 Anos'].astype(float)
df_temp['55 Anos e +'] = df_temp['55 Anos e +'].astype(float)
df_temp['TOTAL'] = df_temp['TOTAL'].astype(float)

df_unemployment_2=df_temp.groupby(['concelhos']).mean().astype(int).reset_index()
#df_unemployment_2=df_unemployment_2[df_unemployment_2['concelhos'].isin(county.unique())].reset_index()

print('Sept 2019 - February 2020')
print(df_unemployment_1.sum(axis = 0, skipna = True))


print('Sept 2020 - January 2021')
print(df_unemployment_2.sum(axis = 0, skipna = True))
```
