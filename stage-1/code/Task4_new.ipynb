```python
import json
import pandas as pd
import numpy as np
import plotly.express as px
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly import graph_objs as go
```

```python
#Open files
f = open('concelhos-shapefile.json',)
counties = json.load(f)

df = pd.read_csv("data_NAME_2.csv", dtype={"ID_2": str})
```

```python
counties["features"][0]["properties"]["NAME_2"]
```

```python
#Map
fig = px.choropleth_mapbox(data_frame=df, geojson=counties, color='relative',
                           locations='NAME_2', featureidkey="properties.NAME_2",
                           color_continuous_scale="Viridis",
                           range_color=(df["relative"].min(), df["relative"].max()),
                           center = {"lat": 38.736946, "lon": -9.142685}, 
                           opacity=0.5,
                           labels={"relative":"Relative_Change_AverageBusUsers"},
                           mapbox_style="carto-positron", zoom=9)


fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()    
```
