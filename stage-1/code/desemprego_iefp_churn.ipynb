```python
from pandas_ods_reader import read_ods
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#unemployment dataframe creation for sep 2019 - Feb 2020
mes=['setembro','outubro','novembro','dezembro','Janeiro','fevereiro']
dfn=[]
for idx,i in enumerate(mes):
    if i=='setembro'or i=='outubro' or i=='novembro'or i=='dezembro':
        ano=2019
        path = "dados_iefp/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 4
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,3,300])
        df = df.drop(columns=["unnamed.2","unnamed.8","unnamed.9","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': 'homens_desempregados', 'unnamed.6': 'mulheres_desempregadas', 'unnamed.7': 'total_desempregados'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn.append(df)
    else:
        ano=2020
        path = "dados_iefp/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 4
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,3,300])
        df = df.drop(columns=["unnamed.2","unnamed.8","unnamed.9","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': 'homens_desempregados', 'unnamed.6': 'mulheres_desempregadas', 'unnamed.7': 'total_desempregados'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn.append(df)

        
df_final=pd.concat(dfn[i] for i in range(6))
df_final['homens_desempregados'] = df_final['homens_desempregados'].astype(float)
df_final['mulheres_desempregadas'] = df_final['mulheres_desempregadas'].astype(float)
df_final['total_desempregados'] = df_final['total_desempregados'].astype(float)
df_desemprego_1=df_final.groupby(['concelhos']).mean().astype(int).reset_index()


#unemployment dataframe creation for sep 2020 - January 2021
mes=['setembro','outubro','novembro','dezembro','Janeiro']
dfn=[]
for idx,i in enumerate(mes):
    if i=='setembro'or i=='outubro' or i=='novembro'or i=='dezembro':
        ano=2020
        path = "dados_iefp/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 4
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,3,300])
        df = df.drop(columns=["unnamed.2","unnamed.8","unnamed.9","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': 'homens_desempregados', 'unnamed.6': 'mulheres_desempregadas', 'unnamed.7': 'total_desempregados'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn.append(df)
    else:
        ano=2021
        path = "dados_iefp/SIE - Desemprego registado por concelhos %s %d.ods" %(i,ano)
        # Data cleaning
        sheet_idx = 4
        df = read_ods(path, sheet_idx)
        df=df.drop([0,1,2,3,300])
        df = df.drop(columns=["unnamed.2","unnamed.8","unnamed.9","unnamed.10"])
        df = df.rename(columns = {'unnamed.4': 'concelhos', 'unnamed.5': 'homens_desempregados', 'unnamed.6': 'mulheres_desempregadas', 'unnamed.7': 'total_desempregados'}, inplace = False)
        df['concelhos']=df['concelhos'].str.lower()
        df=df.dropna()
        dfn.append(df)

df_final=pd.concat(dfn[i] for i in range(5))
df_final['homens_desempregados'] = df_final['homens_desempregados'].astype(float)
df_final['mulheres_desempregadas'] = df_final['mulheres_desempregadas'].astype(float)
df_final['total_desempregados'] = df_final['total_desempregados'].astype(float)
df_desemprego_2=df_final.groupby(['concelhos']).mean().astype(int).reset_index()
```
