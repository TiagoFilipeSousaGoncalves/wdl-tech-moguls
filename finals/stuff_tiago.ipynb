```python
def gauss_to_wsg84(Eutm, Nutm):
    """
    Converts Gauss-Boaga coordinates to WGS84 (lat, long) coordinates
    Source: https://stackoverflow.com/questions/24169285/gauss-boaga-to-wgs84
    
    """
    
    # Parameter that depend on the fuso ovest
    # $l0 = 9;
    l0 = 9
    
    # Another parameter we obtain
    # $fraz = $Nutm/111092.0821;
    fraz = Nutm / 111092.0821
    
    
    # $A= $fraz + (0.1449300705 * sin(deg2rad(2*$fraz))) + (0.0002138508 * sin(deg2rad(4*$fraz))) + (0.0000004322 * sin(deg2rad(6*$fraz)));
    A = fraz + (0.1449300705 * math.sin(math.radians(2*fraz))) + (0.0002138508 * math.sin(math.radians(4*fraz))) + (0.0000004322 * math.sin(math.radians(6*fraz)))
    
    
    # $v = sqrt(1 + (0.0067681702 * cos(deg2rad($A)) * cos(deg2rad($A)) ) );
    v = math.sqrt(1 + (0.0067681702 * math.cos(math.radians(A)) * math.cos(math.radians(A))))
    
    # $y = $Eutm - 500000;
    y = Eutm - 500000
    
    # $B = rad2deg(atan( ($v * sinh(deg2rad($y/6397376.633)) ) / cos(deg2rad($A))));
    B = math.degrees(math.atan((v * math.sinh(math.radians(y/6397376.633))) / math.cos(math.radians(A))))

    # $lng = rad2deg(atan(tan(deg2rad($A)) * cos(deg2rad($v * $B))));
    long = math.degrees(math.atan(math.tan(math.radians(A)) * math.cos(math.radians(v * B))))
    
    # $lat = $B + $l0;
    lat = B + l0
    
    
    return long, lat
```

```python
# Create a Geolocator
geolocator = Nominatim(user_agent="wdl-tech-moguls")

# Note Google Maps reads (lat, long)
# iterate through each row and select 
# 'Name' and 'Age' column respectively.
for i in range(len(gdf)):
    location = geolocator.geocode(gdf.loc[i, "ADDRESS"])
    gdf.loc[i, "Longitude"], gdf.loc[i, "Latitude"] = location.longitude, location.latitude

gdf
```

```python
final_businesses = gdf.copy()[["WKT", "ADDRESS","OPEN YEAR", "OPEN MONTH", "TYPE", "Description", "Merchandise Type", "Longitude", "Latitude"]]
final_businesses
```
